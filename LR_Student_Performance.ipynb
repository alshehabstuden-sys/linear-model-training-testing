```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data = pd.read_csv(r"D:\تقنيات الذكاء الاصطناعي\التكاليف+المشروع\Final project\score_updated.csv") # real
data.tail()
data.info()
data.describe()
```

```python
plt.figure(figsize=(12,6))
sns.pairplot(data,x_vars=["Hours"],y_vars=["Scores"],size=7,kind="scatter")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Scores prediction")
plt.show()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,1]

X.head()
y.head()

X_train , X_test , y_train , y_test = train_test_split(X,y,train_size=0.8,random_state = 10)
my_model = LinearRegression()
my_model.fit(X_train,y_train)
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, my_model.predict(X_train), color = 'blue')
plt.title('Scores vs Hours (Training set)')
plt.xlabel('Hours ')
plt.ylabel('Scores')
plt.show()
```

```python
y_pred = my_model.predict(X_test)
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Scores vs Hours (Test set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
c = [i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test,color='r',linestyle='-')
plt.plot(c,y_pred,color='b',linestyle='-')
plt.xlabel('Scores')
plt.ylabel('index')
plt.title('Prediction')
plt.show()
```

```python
c = [i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test-y_pred,color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
rsq = r2_score(y_test,y_pred)
print('mean squared error :',mse)
print('r square :',rsq)
```

```python
print('Intercept of the model:',my_model.intercept_)
print('Coefficient of the line:',my_model.coef_)
```
